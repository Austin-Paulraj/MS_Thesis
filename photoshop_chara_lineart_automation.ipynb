```python
def collect_data(ds_path):
    ds = {"path":[], "filename":[], "image_name":[]}
    for source in os.listdir(ds_path):
        #print(source)
        source_path = os.path.join(ds_path,source)
        
        for character in os.listdir(source_path):
            #print(character)
            character_path = os.path.join(source_path,character)
            
            ds["path"].append(character_path)
            file_aux = []
            img_aux = []
            for art_type in os.listdir(character_path):
                #print(art_type)
                if art_type == "Concept_Art":
                    art_path = os.path.join(character_path,art_type)
                    
                    
                    

                    for image_filename in os.listdir(art_path):
                        file_aux.append(os.path.join(art_type,image_filename))
                        img_aux.append(image_filename)
                
            ds["filename"].append(file_aux)
            ds["image_name"].append(img_aux)
    return ds
```

```python
def lineart_action(app):
    idPly = app.charIDToTypeID("Ply ")
    desc8 = ps.ActionDescriptor()
    idnull = app.charIDToTypeID("null")

    ref3 = ps.ActionReference()
    idActn = app.charIDToTypeID("Actn")
    ref3.putName(idActn, "Prep_chara_lineart")
    idASet = app.charIDToTypeID("ASet")
    ref3.PutName(idASet, "Default Actions")
    desc8.putReference(idnull, ref3)
    app.executeAction(idPly, desc8, ps.DialogModes.DisplayNoDialogs)

def bg_remove_action(app):
    idPly = app.charIDToTypeID("Ply ")
    desc8 = ps.ActionDescriptor()
    idnull = app.charIDToTypeID("null")

    ref3 = ps.ActionReference()
    idActn = app.charIDToTypeID("Actn")
    ref3.putName(idActn, "remove_bg")
    idASet = app.charIDToTypeID("ASet")
    ref3.PutName(idASet, "Default Actions")
    desc8.putReference(idnull, ref3)
    app.executeAction(idPly, desc8, ps.DialogModes.DisplayNoDialogs)

def show_mask_action(app):
    idPly = app.charIDToTypeID("Ply ")
    desc8 = ps.ActionDescriptor()
    idnull = app.charIDToTypeID("null")

    ref3 = ps.ActionReference()
    idActn = app.charIDToTypeID("Actn")
    ref3.putName(idActn, "show_mask")
    idASet = app.charIDToTypeID("ASet")
    ref3.PutName(idASet, "Default Actions")
    desc8.putReference(idnull, ref3)
    app.executeAction(idPly, desc8, ps.DialogModes.DisplayNoDialogs)

def adjust_la_action(app):
    idPly = app.charIDToTypeID("Ply ")
    desc8 = ps.ActionDescriptor()
    idnull = app.charIDToTypeID("null")

    ref3 = ps.ActionReference()
    idActn = app.charIDToTypeID("Actn")
    ref3.putName(idActn, "adjust_line_art")
    idASet = app.charIDToTypeID("ASet")
    ref3.PutName(idASet, "Default Actions")
    desc8.putReference(idnull, ref3)
    app.executeAction(idPly, desc8, ps.DialogModes.DisplayNoDialogs)

def readjust_mask_action(app):
    idPly = app.charIDToTypeID("Ply ")
    desc8 = ps.ActionDescriptor()
    idnull = app.charIDToTypeID("null")

    ref3 = ps.ActionReference()
    idActn = app.charIDToTypeID("Actn")
    ref3.putName(idActn, "readjust_mask")
    idASet = app.charIDToTypeID("ASet")
    ref3.PutName(idASet, "Default Actions")
    desc8.putReference(idnull, ref3)
    app.executeAction(idPly, desc8, ps.DialogModes.DisplayNoDialogs)

def sep_and_trim_action(app):
    idPly = app.charIDToTypeID("Ply ")
    desc8 = ps.ActionDescriptor()
    idnull = app.charIDToTypeID("null")

    ref3 = ps.ActionReference()
    idActn = app.charIDToTypeID("Actn")
    ref3.putName(idActn, "sep_and_trim")
    idASet = app.charIDToTypeID("ASet")
    ref3.PutName(idASet, "Default Actions")
    desc8.putReference(idnull, ref3)
    app.executeAction(idPly, desc8, ps.DialogModes.DisplayNoDialogs)

def readjust_mask_2_action(app):
    idPly = app.charIDToTypeID("Ply ")
    desc8 = ps.ActionDescriptor()
    idnull = app.charIDToTypeID("null")

    ref3 = ps.ActionReference()
    idActn = app.charIDToTypeID("Actn")
    ref3.putName(idActn, "readjust_mask_2")
    idASet = app.charIDToTypeID("ASet")
    ref3.PutName(idASet, "Default Actions")
    desc8.putReference(idnull, ref3)
    app.executeAction(idPly, desc8, ps.DialogModes.DisplayNoDialogs)
```

```python
def extract_fullimage_features(path, filename, image_name):
    app = ps.Application()
    doc = app.load(os.path.join(path,filename))
    save_folder = os.path.join("C:/Austin/Projects/MS_Thesis/test",image_name)
    if not os.path.exists(save_folder):



        if not os.path.exists(save_folder):
            os.makedirs(save_folder)

        options = ps.PNGSaveOptions()
        ps_options = ps.PhotoshopSaveOptions()


        if not os.path.exists(os.path.join(save_folder,"Full_images")):
            os.makedirs(os.path.join(save_folder,"Full_images"))
        
        if not os.path.exists(os.path.join(save_folder,"Extracted_images")):
            os.makedirs(os.path.join(save_folder,"Extracted_images"))

        doc.saveAs(os.path.join(save_folder,"Full_images/original_image.png"), options, asCopy=True)
        
        lineart_action(app)
        bg_remove_action(app)

        input("Fix Mask")

        layers = doc.artLayers
        for layer in layers:
            layer.visible = False

        readjust_mask_action(app)

        layers[0].visible = True
        doc.saveAs(os.path.join(save_folder,"Full_images/bg_mask.png"), options, asCopy=True)
        layers[0].visible = False


        layers[4].visible = True
        layers[2].visible = True
        adjust_la_action(app)
        doc.saveAs(os.path.join(save_folder,"Full_images/bg_removed_image.png"), options, asCopy=True)
        layers[4].visible = False
        layers[2].visible = False

        doc.saveAs(os.path.join(save_folder,"Full_images/full_image.psd"), ps_options, True)

        doc.close()

    return os.path.join(save_folder,"Full_images/full_image.psd")

        #input("done?")




    
    

```

```python
def extract_character_features(target_path, i, image_name):
        options = ps.PNGSaveOptions()
        app = ps.Application()
        doc = app.load(target_path)

        layers = doc.artLayers
        for layer in layers:
                layer.visible = False
        
        layers[0].visible = True

        save_folder = os.path.join("C:/Austin/Projects/MS_Thesis/test",image_name)
        char_folder = os.path.join(save_folder,"Extracted_images")
        full_folder = os.path.join(save_folder,"Full_images")

        save_type = input("Type of extract")
        char_folder = os.path.join(char_folder, str(i)+"_"+save_type)
        full_folder = os.path.join(full_folder, str(i)+"_"+save_type)
        if not os.path.exists(char_folder):
                os.makedirs(char_folder)

        if not os.path.exists(full_folder):
                os.makedirs(full_folder)

        input("Select")

        sep_and_trim_action(app)

        layers = doc.artLayers
        for layer in layers:
                layer.visible = False

        layers[0].visible = True
        doc.saveAs(os.path.join(full_folder,"bg_mask.png"), options, asCopy=True)
        layers[0].visible = False


        layers[4].visible = True
        layers[2].visible = True
        doc.saveAs(os.path.join(full_folder,"bg_removed_image.png"), options, asCopy=True)

        doc.trim(ps.TrimType.TopLeftPixel, True, True, True, True)

        layers = doc.artLayers
        for layer in layers:
                layer.visible = False

        layers[0].visible = True
        doc.saveAs(os.path.join(char_folder,"bg_mask.png"), options, asCopy=True)
        layers[0].visible = False


        layers[4].visible = True
        layers[2].visible = True
        doc.saveAs(os.path.join(char_folder,"bg_removed_image.png"), options, asCopy=True)
        layers[4].visible = False
        layers[2].visible = False

        doc.close()
```
