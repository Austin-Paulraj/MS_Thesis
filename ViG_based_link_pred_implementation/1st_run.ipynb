```python
with open('/home/paulraae/MS_Thesis_Data/tester_eye_data/test_eye_clusters_extract_202.pk1', 'rb') as handle:
    data = pickle.load(handle)
```

```python
def draw_grid(image, n=2):
    ln = int(224/n)
    image = cv2.resize(image,(ln*n, ln*n))
    
    H = ln
    W = ln
    im2 = copy.deepcopy(image)
    tiles = [im2[x:x+H,y:y+W] for x in range(0,im2.shape[0],H) for y in range(0,im2.shape[1],W)]
    
    return np.array(tiles)

in_image = data["image"][0]
tiles = draw_grid(in_image, nt)
print(tiles.shape)
```

```python
data["tiles"] = []
for img in data["image"]:
    data["tiles"].append(draw_grid(img,nt))

data["tiles"] = np.array(data["tiles"])
```
